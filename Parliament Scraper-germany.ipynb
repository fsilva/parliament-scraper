Sessions are listed in:
https://www.bundestag.de/dokumente/protokolle/plenarprotokolle/plenarprotokolle

but it's a fully javascript page. Links are added by javascript and only available after page-load, so its not just parsing the DOM. Also, in order to go the the next set of sessions one needs to click a button.


We can automate all of this.


Sessions from 2013 to 2017 in txt format.


```python
url = 'https://www.bundestag.de/dokumente/protokolle/plenarprotokolle/plenarprotokolle';

import sys  

from PyQt5.QtGui import *  
from PyQt5.QtCore import *  
from PyQt5.QtWebEngineWidgets import *
from PyQt5.QtWidgets import QApplication
from lxml import html 
import bs4
from threading import Timer

i = 0
germany_data = []

def process_page(html):
    global i, germany_data, soup
    soup = bs4.BeautifulSoup(html,'html.parser')
    elem = soup.findAll('a',{'class','bt-link-dokument'})
    for e in elem:
        page_url = 'https://www.bundestag.de'+e['href'];
        germany_data.append(page_url)
        print(str(i)+ " "+page_url)
        i= i+1
    import json

    with open('parliament-url-germany.json','w') as f:
        json.dump(germany_data,f)
        

class view(QWebEngineView):
    def __init__(self):
        print('here1')
        QWebEngineView.__init__(self)
        self.loadFinished.connect(self._result_available)

    def _result_available(self, ok):
        print('here2')
        frame = self.page()
        frame.toHtml(process_page)
        
        def fun(frame):
            js = 'document.querySelector(\'.slick-next.slick-arrow\').click();'
            print(js)
            #js = 'document.querySelectorAll("button")[-1].click();'
            frame.runJavaScript(js,lambda a: print('ran' + str(a)))
        

        t = Timer(3.0, fun, [frame])
        t.start()
        
        t = Timer(6.0, self._result_available, [self])
        t.start()

app = QApplication(sys.argv)
wv = view()
wv.load(QUrl(url))
wv.show()
print('here3')
app.exec_()
```

```python
import requests
import time
import re
from datetime import datetime
import locale

locale.setlocale(locale.LC_TIME, "de_DE")

germany_data = []
i = 0
start = time.time()
for page_url in germany_urls:
    page = requests.get(page_url);
    #print(page.encoding)
    print(page_url)
    tmp = page.content #load encoding
    page.encoding = 'UTF-8'

    text = page.content.decode('utf-8','ignore')
    
    s = re.search('[0-9].*Januar.*\d\d\d\d|[0-9].*Februar.*\d\d\d\d|[0-9].*Marz.*\d\d\d\d|[0-9].*M.*rz.*\d\d\d\d|[0-9].*April.*\d\d\d\d|[0-9].*Mai.*\d\d\d\d|[0-9].*Juni.*\d\d\d\d|[0-9].*Juli.*\d\d\d\d|[0-9].*August.*\d\d\d\d|[0-9].*September.*\d\d\d\d|[0-9].*Oktober.*\d\d\d\d|[0-9].*November.*\d\d\d\d|[0-9].*Dezember.*\d\d\d\d',text).group(0).strip().replace('.','')
#    print(s)
    
#    s = s.encode('ascii',errors='ignore').decode('ascii')
    s = s.replace('Mrz','März')
    
    datet = datetime.strptime(s, '%d %B %Y')
    date = datet.strftime("%Y-%m-%d")
    print(s + ' = ' + date+ " "+page_url)
    
    germany_data.append({'date':date,'url':page_url,'text':text})
    i = i + 1
    percentage = i*100/len(germany_urls)
    time_elapsed = time.time()-start
    time_left = time_elapsed/percentage*100
    print("{0:.2f}".format(percentage)+"% "+date+ " " + "{0:.2f}".format(time_elapsed) + " s elapsed - " + "{0:.2f}".format(time_left) + " s total")
```
