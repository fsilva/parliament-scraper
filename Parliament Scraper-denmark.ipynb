==========

Sessions are listed in:
http://www.ft.dk/Dokumenter/Vis_efter_type/Referater_efter_modedato.aspx#dok
in HTML form. 1030 sessions from 2007 to 2017.


```python
import requests
import bs4

i = 0
denmark_pages = []
for year in range(2005,2018):
#for year in range(2005,2008):
    for month in range(1,2):
        #url = 'http://www.ft.dk/dokumenter/vis_efter_type/referater_efter_modedato.aspx?session=&meetingNumber=&startDate='+str(year)+"{0:0=2d}".format(month)+'01&endDate='+str(year)+"{0:0=2d}".format(month)+'31&sortColumn=&sortOrder=&startRecord=1&numberOfRecords=100&totalNumberOfRecords=#dok'
        url = 'http://www.ft.dk/dokumenter/vis_efter_type/referater_efter_modedato.aspx?session=&meetingNumber=&startDate='+str(year)+'0101&endDate='+str(year)+'1231&sortColumn=&sortOrder=&startRecord=1&numberOfRecords=1000&totalNumberOfRecords=#dok'
        page = requests.get(url);

        soup = bs4.BeautifulSoup(page.text,'html.parser')


        for element in soup.findAll('tr'):
            try:
                page = 'http://www.ft.dk'+element['onclick'].split('\'')[1]
                if(page[-3:] == 'htm'):
                    denmark_pages.append(page)
                    print(str(i)+" "+page)
                    i = i + 1
            except: 
                pass

```

```python
len(denmark_pages)
```

```python
import re 
import time

denmark_data = []
i = 0
start = time.time()
for page_url in denmark_pages:
    try:
        page = requests.get(page_url);
        #print(page.encoding)
        tmp = page.content #load encoding
        page.encoding = 'UTF-8'

        soup = bs4.BeautifulSoup(page.text,'html.parser')
        date = str(soup.find('meta',{"name":"DateOfSitting"})['content'])
        date = date[:10]+' '+date[11:]
        
        [x.decompose() for x in soup.findAll('p',{'class':'Tid'})]
        [x.decompose() for x in soup.findAll('meta')]
        [x.decompose() for x in soup.findAll('link')]
        [x.decompose() for x in soup.findAll('hr')]
        text = ''
        for tag in soup.find('body').contents:
            text = text + str(tag)
        text = re.sub('<[^>]*>', '', text) #remove remaining tags

        text = text.replace(' \n','\n')
        text = text.replace('\n\n\n\n','\n')
        text = text.replace('\n\n\n','\n')
        text = text.replace('\n\n','\n')
    
        denmark_data.append({'date':date,'url':page_url,'text':text})
        i = i + 1
        percentage = i*100/len(denmark_pages)
        time_elapsed = time.time()-start
        time_left = time_elapsed/percentage*100
        print("{0:.2f}".format(percentage)+"% "+date+ " " + "{0:.2f}".format(time_elapsed) + " s elapsed - " + "{0:.2f}".format(time_left) + " s total")

    except:
            pass

```

```python
denmark_data = sorted(denmark_data, key=lambda a:a['date'])
```

```python
import json

with open('parliament-scraper-denmark.json','w') as f:
    json.dump(denmark_data,f)
    
```
