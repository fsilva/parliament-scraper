This is extremely slow (10h>)

```python
import requests
import bs4
import os
import re

url = 'http://www.congreso.es'
page = '/portal/page/portal/Congreso/Congreso/Iniciativas?_piref73_2148295_73_1335437_1335437.next_page=/wc/servidorCGI&CMD=VERLST&BASE=IWSE&FMT=INTTXLVS.fmt&DOCS=1-26&DOCORDER=FIFO&OPDEF=Y&QUERY=%40FDIA%26gt%3B%3D19770101+%26+%40FDIA%26lt%3B%3D20170311'
next_link = url+page

folder = 'spain/'
pageN = 1

visitedLinks = []
datesScraped = []

while(True):
    try:
        print("Scraping page",pageN,'(%.2f'%(pageN*100/13038)+"%)","...")
        pageN += 1
        page = requests.get(next_link);
        soup = bs4.BeautifulSoup(page.text,'html.parser')
        next_links = soup.find(class_ = 'paginacion_brs').findAll('a')
        # check if its last
        last_a = None
        thereIsMore = False
        for last_a in next_links:
            if last_a.text == str(pageN+1):
                next_link = url + last_a['href']
                thereIsMore = True
        if not thereIsMore:
            break

        links = []
        for link in soup.findAll('a', href=True):
            if 'ntegro' in link.text:
                new_link = (url+link['href']).rsplit("#",1)[0]
                if new_link not in links:
                    links.append(new_link)
        count = 0
        for link in links:
            if link not in visitedLinks:
                #print(link)
                page = requests.get(link);
                soup = bs4.BeautifulSoup(page.content.decode('utf-8','ignore'),'html.parser')
                try:
                    match = re.search(r'(\d+/\d+/\d+)',page.text)
                    date = match.group(1)
                    if date not in datesScraped:
                        datesScraped.append(date)
                        print("Scraped",date)
                        with open(folder + date.replace("/","-") + ".txt","w", encoding="utf-8") as a:
                            a.write(soup.find(id='CONGRESO_TEXTO_INTEGRO_POPUP').text)
                            count += 1
                except:
                    print("failed at:",link)
                visitedLinks.append(link)
    except:
        print("catastrophic failure!")
        print("stopped at:",next_link)
        break

print("downloaded",count,"sessions!")
```
